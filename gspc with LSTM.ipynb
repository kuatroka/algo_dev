```python

seq_length = 5

dataX = []
dataY = []
for i in range(0, len(gspc4["pct_change"].values) - seq_length, 1):
    seq_in = gspc4["pct_change"].values[i:i + seq_length]
    scaler = MinMaxScaler(feature_range=(-1, 1))
    seq_in = scaler.fit_transform(seq_in)

    seq_out = gspc4["pct_change"].values[i + seq_length]
    dataX.append([char for char in seq_in])
    dataY.append([seq_out])
    print(seq_in, '->', seq_out)
```

```python
X = np.reshape(dataX, (len(dataX), seq_length, 1))
# X = X / float(len(gspc4["pct_change"].values))
len(dataY)

# dataY conveeted into binary classification 1.0 and 0.0
dataYseries = pd.DataFrame(dataY)
dataYseries.ix[dataYseries[0] >= 0, ] = 1
dataYseries.ix[dataYseries[0] < 0, ] = 0
X[:20]
```

```python
# moving dataYseries to on-hot-encoder
y = keras.utils.to_categorical(dataYseries)
y
```

## Very important - data needs to be normalized in format [-1, 1] for LSTM
## After this change the results in training accuracy went drastically up

```python
"""
stateful = False
"""
# create and fit the model
batch_size = X.shape[0]
model = keras.models.Sequential()
model.add(keras.layers.LSTM(128, input_shape=(X.shape[1], X.shape[2])))
model.add(keras.layers.Dense(y.shape[1], activation='softmax'))
adam = keras.optimizers.Adam(lr=0.05, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model.fit(X, y, epochs=600, batch_size=batch_size, verbose=1, shuffle=False)

# seq_length = 4
# max accuracy = 0.5419 normalized 0 to 1 with all sorts of lr= [0.0001, 0.001, 0.01]
# max accuracy 0.5492 with normalized between -1 and 1, lr= 0.01
# neur = 3, epochs=200, bs = 1, norm [-1, 1], lr=0.05, acc 0.58
# neur = 3, epochs=200, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.58
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.59
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.04 --  acc = 0.584
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.06 --  acc = 0.5953
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.07 --  acc = 0.5953
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.08 --  acc = 0.5975
# neur = 3, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.1 --  acc = 0.5840
# neur = 16, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.747
# neur = 32, epochs=400, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.81
# neur = 64, epochs=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.85
# neur = 128, epochs=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.92
# seq = 5, neur = 128, epochs=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 1.0
```

```python
"""
stateful = True
"""
# create and fit the model
batch_size = X.shape[0]
model = keras.models.Sequential()
model.add(keras.layers.LSTM(16, batch_input_shape=(batch_size, X.shape[1], X.shape[2]), stateful=True))
model.add(keras.layers.Dense(y.shape[1], activation='softmax'))
adam = keras.optimizers.Adam(lr=0.05, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0)
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
for i in range(600):
	model.fit(X, y, epochs=1, batch_size=batch_size, verbose=2, shuffle=False)
	model.reset_states()
# summarize performance of the model
scores = model.evaluate(X, y, batch_size=batch_size, verbose=0)
model.reset_states()
print("Model Accuracy: %.2f%%" % (scores[1]*100))

# the same accuracy as in stateful=False  = 0.5419 !! with input data being normalized between 0 and 1
# after changing it to be between -1 and 1 , accuracy went up to 0.5554
# after increasing lr from 0.01 to 0.05, accuracy = 0.73
# seq = 4, neur = 32, epo=300, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.856
# seq = 4, neur = 64, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.87
# seq = 4, neur = 128, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.946
# seq = 4, neur = 256, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.648
# seq = 4, neur = 128, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.616
# seq = 5, neur = 128, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 1.0
# seq = 5, neur = 64, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 1.0
# seq = 5, neur = 32, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 1.0
# seq = 5, neur = 16, epo=600, bs = X.shape[0], norm [-1, 1], lr=0.05 --  acc = 0.90
```
