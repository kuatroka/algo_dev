```python
import datetime as datetime
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import numpy as np
# do all our graphics inline
%matplotlib inline
```

```python
start = datetime.date(2016, 1, 1)
end = datetime.date(2016, 9, 14)

# get data for MSFT
nugt = web.DataReader('NUGT', "yahoo", start, end)[['Adj Close']]
nugt.reset_index(inplace=True)
```

```python
nugt['rollmax'] = pd.rolling_max(nugt['Adj Close'], 8)
nugt['rollmin'] = pd.rolling_min(nugt['Adj Close'], 8)
```

```python
# here we use the lookup logic to create a new column 'rollmax_date'. the logic is:
# 'rollmax_date' = 'Date' where 'Adj Close' == 'rollmax'
#(nugt.loc[nugt['rollmax'] == nugt['Adj Close']])
#nugt['rollmax_date'] = nugt["Date"][(nugt["Adj Close"] == nugt.rollmax) & (df["C"] == 900)]
#nugt['rollmax_date'] = np.where(nugt['Adj Close'] == nugt.rollmax, 'Today is MAX', '0'); nugt
nugt['rollmax_date'] = nugt.loc[(nugt["Adj Close"] == nugt.rollmax), 'Date']
nugt
```

```python
nugt = nugt[['Date', 'Adj Close', 'rollmax', 'rollmin', 'rollmax_date']]
nugt
```
